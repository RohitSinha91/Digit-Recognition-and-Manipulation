```python
#defining threshold which will convert colored images into a b&w image
from statistics import mean

def threshold(imageArray) :
    balanceAr = []
    newAr = imageArray

    for eachRow in imageArray:
        for eachPix in eachRow:
            avgNum = mean(eachPix[:3])        #this will give the avg of RGB
            balanceAr.append(avgNum)

    balance = mean(balanceAr)                #we appended that in balanceAr, now we are finding out the mean of that
    for eachRow in newAr:                   #coz you dont mess with numpy arrays
        for eachPix in eachRow:
            if mean(eachPix[:3]) < balance:
                eachPix[0] = 255
                eachPix[1] = 255
                eachPix[2] = 255
                eachPix[3] = 255
            else:
                eachPix[0] = 0
                eachPix[1] = 0
                eachPix[2] = 0
                eachPix[3] = 255
    
    return newAr
threshold(iar3)
threshold(iar2)
threshold(iar4)

fig = plt.figure()
ax1 = plt.subplot2grid((8,6), (0,0), rowspan = 4, colspan = 3)
ax2 = plt.subplot2grid((8,6), (4,0), rowspan = 4, colspan = 3)
ax3 = plt.subplot2grid((8,6), (0,3), rowspan = 4, colspan = 3)
ax4 = plt.subplot2grid((8,6), (4,3), rowspan = 4, colspan = 3)

ax1.imshow(iar)
ax2.imshow(iar2)
ax3.imshow(iar3)
ax4.imshow(iar4)

plt.show() 
```

```python
#create a function to take a current image and then compare it to the database we just made with all the numbers
from collections import Counter
def whatNumIsThis(filePath):
    matchedAr = []              #creating an empty array
    loadExamps = open('numArEx.txt', 'r').read()    #reading the previous file in
    loadExamps = loadExamps.split('\n')            #splitting it by every line
    
    i = Image.open(filePath)              #loading the file specified by user
    iar = np.array(i)               #converting it into an array
    iarl = iar.tolist()                 #converting into list for better storage
    
    inQuestion = str(iarl)            
    
    for eachExample in loadExamps:         #
        if len(eachExample) > 3:
            splitEx = eachExample.split('::')
            currentNum = splitEx[0]                   #this contains the part before ::
            currentAr = splitEx[1]                #this contains the part after :: which is the actual array for the picture
            
            eachPixEx = currentAr.split('],')    #loading the pixel values from our file numArEx in the variable
            
            eachPixInQ = inQuestion.split('],')   #loading the pixel values of the question in the variable
            
            x = 0
            
            while x < len(eachPixEx) :
                if eachPixEx[x] == eachPixInQ[x] :         #comparing pixels of question and the value in our file    
                    matchedAr.append(int(currentNum))
                    
                x += 1
                
    #print(matchedAr)
    x = Counter(matchedAr)
    
    print(x)
      
    graphX = []               #empty array
    graphY = []               #empty array
    
    for eachThing in x :
        #print(eachThing)
        graphX.append(eachThing)          #giving values for x axis
        #print(x[eachThing])
        
        graphY.append(x[eachThing])      #giving values for y axis
        
    
    fig = plt.figure()
    ax1 = plt.subplot2grid((4,4),(0,0), rowspan = 1, colspan=4)
    ax2 = plt.subplot2grid((4,4),(1,0), rowspan = 3, colspan=4)
    
    ax1.imshow(iar)
    ax2.bar(graphX, graphY, align = 'center')
    
    plt.ylim(380)
   
    
    xloc = plt.MaxNLocator(12)           #increasing the label count on x axis
    
    ax2.xaxis.set_major_locator(xloc)
    
    plt.show()
    
whatNumIsThis('C:/Users/Rohit/Documents/Image recognition Python/tutorialimages/images/test.png')   
      
    
```
